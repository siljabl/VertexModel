```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Experimental parameters

v_exp   = 1    # µm/frame
V_exp   = 1700 # µm³
tau_exp = 0.75 *12 # frame

# Model parameters

V0   = 4.5  # r_6
taup = 1    

# Conversion
um_to_r6  = (V0 / V_exp)**(1/3)
h_to_taup = taup / tau_exp

v_exp * um_to_r6 / h_to_taup
```

```python
# Correlation

t = np.linspace(0, 10, 100)
def corr(t, v0, taup): return v0**2 * np.exp(-t / taup)
```

```python
plt.axhline(0, ls="dashed")
plt.plot(t, corr(t, 1, 1))
plt.plot(t/0.5, corr(t, 1, 0.5))
plt.plot(t/0.75, corr(t, 1, 0.75))
```
